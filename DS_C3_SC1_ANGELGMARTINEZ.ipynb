
Guarda en un dataframe (happiness) todos los registros del archivo Happiness_report.csv.


```python
import pandas as pd

happiness = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/0.rw6r6mdtjng0.9sxn8dzeqsbHappiness_report.csv')

happiness
```

A partir del dataframe happiness obtén otro (mexico) donde únicamente almacenes la información de México. Haz que la columna Country or region quede como índice.

```python
#Creamos nuestro dataframe (mexico) a partir de 'happiness'
mexico = happiness[happiness['Country or region'] == 'Mexico']

#Hacemos que la columna 'Country or region' quede como el indice
mexico.set_index('Country or region', inplace = True)

mexico
```

```python
#Como un plus cambiamos hacemos que la columna 'Country or region' quede como indice de todo el dataframe principal, lo cual nos facilitará posteriores analisis
happiness.set_index('Country or region', inplace = True)

happiness
```

Obtén un histograma del puntaje (score) para identificar el rango o clase más frecuente. Ubica el valor de México con una etiqueta de texto.

Se personalizan los elementos del gráfico (título y etiquetas de los ejes).

```python
#Hacemos el histograma del score con matplotlib
import matplotlib.pyplot as plt

plt.hist(happiness['Score'], color='mediumaquamarine')
plt.xlabel('Score')
plt.ylabel('Frequency')
plt.title('World Happiness Report')


#Creamos diccionarios con las especificaciones esteticas de nuestra anotacion y texto
prop_box = {'boxstyle': 'circle', 'facecolor': 'red', 'alpha': 0.3}

font_box = {'color': 'white', 'weight': 'bold', 'size': 10}

#Usamos nuestro dataframe anterior 'mexico' para determinar donde colocaremos la anotacion "mex" en un circulo rojo dentro del histograma
datox_mex = mexico.loc['Mexico', 'Score']

plt.text(datox_mex, 0, 'MEX',
        bbox = prop_box,
        fontdict = font_box)

plt.show()
```

Construye un boxplot para la columna esperanza de vida (Healthy life expectancy). Ubica el valor de México con una anotación.

```python
#hacemos nuestro boxplot para 'Healthy life expectancy' con colores mediumaquamarine
import matplotlib.pyplot as plt

plt.boxplot(happiness['Healthy life expectancy'],
            labels = ['Healthy life expectancy'],
            patch_artist=True,
            boxprops=dict(facecolor='none', color='mediumaquamarine'),
            whiskerprops=dict(color='mediumaquamarine'),
            capprops=dict(color='black'),
            medianprops=dict(color='black'))

plt.title('World Happiness Report')

#hacemos un diccionario para configurar la estetica de la flecha
props_arrow = {'facecolor':'black', 'width':0.5, 'headwidth':4, 'headlength':8}

#Usamos los diccionarios pasados para la etiqueta 'MEX'
prop_box = {'boxstyle': 'circle', 'facecolor': 'red', 'alpha': 0.3}
font_box = {'color': 'white', 'weight': 'bold', 'size': 10}


#Usamos nuestro dataframe anterior 'mexico' para determinar donde colocaremos la anotacion "mex" en un circulo rojo dentro del histograma
mex_healthylifeexpectancy = mexico.loc['Mexico', 'Healthy life expectancy']

#Colocamos la misma anotacion (del ejercicio pasado) 'mex' usando nuestro 'mex_healthylifeexpectancy'
plt.text(1.21, mex_healthylifeexpectancy, 'MEX',
        bbox = prop_box,
        fontdict = font_box)

#colocamos la flecha
plt.annotate('', xy=(1,mex_healthylifeexpectancy), xytext=(1.2,mex_healthylifeexpectancy+0.02), arrowprops=props_arrow,)

plt.show()
```

Filtra el dataframe para quedarte con 5 países (el más feliz, el menos feliz, México y dos más de tu interés) y visualiza en una misma gráfica los 6 factores.

```python
#Seleccionamos el con el mayor y menor 'Score' y lo guardamos en el nuevo dataframe 'Paises'
paises = happiness[(happiness['Score']==happiness['Score'].max()) | (happiness['Score']==happiness['Score'].min()) | (happiness.index=='Mexico') | (happiness.index=='Canada') | (happiness.index=='Italy') ]

paises=paises.T
paises.drop(['Overall rank','Score'], axis = 0, inplace = True)


paises.plot(kind='bar').set_title('Worlds Happiness Report')
```

Crea una matriz de subgráficas de 2 x 3 con scatter plots del puntaje (score) versus los 6 factores, para determinar qué factor influye más en la evaluación.


```python
#Hacemos un nuevo dataframe con los 6 factores a comparar con 'Score'
happiness2 = happiness.drop(['Overall rank'], axis = 1)

#Creamos una figura y subgraficas (2x3)
fig,axs = plt.subplots(2,3,figsize=(15,10))

#Colocamos el titulo de la figura
fig.suptitle('Score vs. Factores del Informe de Felicidad Mundial')

#Graficamos cada relacion y las ubicamos dentro de la figura
axs[0,0].scatter(happiness2['Score'],happiness2['GDP per capita'])
axs[0,0].set_title('GDP per capita vs Score')
axs[0,0].set_xlabel('Score')
axs[0,0].set_ylabel('GDP per capita')

axs[0,1].scatter(happiness2['Score'],happiness2['Social support'])
axs[0,1].set_title('Social support vs Score')
axs[0,1].set_xlabel('Score')
axs[0,1].set_ylabel('Social support')

axs[0,2].scatter(happiness2['Score'],happiness2['Healthy life expectancy'])
axs[0,2].set_title('Healthy life expectancy vs Score')
axs[0,2].set_xlabel('Score')
axs[0,2].set_ylabel('Healthy life expectancy')

axs[1,0].scatter(happiness2['Score'],happiness2['Freedom to make life choices'])
axs[1,0].set_title('Freedom to make life choices vs Score')
axs[1,0].set_xlabel('Score')
axs[1,0].set_ylabel('Freedom to make life choices')

axs[1,1].scatter(happiness2['Score'],happiness2['Generosity'])
axs[1,1].set_title('Generosity vs Score')
axs[1,1].set_xlabel('Score')
axs[1,1].set_ylabel('Generosity')

axs[1,2].scatter(happiness2['Score'],happiness2['Perceptions of corruption'])
axs[1,2].set_title('Perceptions of corruption vs Score')
axs[1,2].set_xlabel('Score')
axs[1,2].set_ylabel('Perceptions of corruption')

plt.show()
```

Combina con el dataframe metada (Metadata.csv) para graficar la felicidad promedio por región.

```python
import pandas as pd
import matplotlib.pyplot as plt


#Leemos el archivo 'Metadata.csv'
metadata = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/0.9ep3krj7s1o0.ona2jvufk4mMetadata.csv')
happiness = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/0.rw6r6mdtjng0.9sxn8dzeqsbHappiness_report.csv')

meta_happiness = pd.merge(left=metadata,right=happiness,left_on='TableName',right_on='Country or region')
meta_happiness.set_index('Country or region', inplace = True)

meta_happiness = meta_happiness.groupby('Region')['Score'].mean()
meta_happiness

#Graficamos la felicidad promedio por region
plt.bar(meta_happiness.index, meta_happiness.values, color='green')
plt.title('Promedio de felicidad por region')
plt.xlabel('Region')
plt.ylabel('Score')
plt.xticks(rotation=90)
plt.show()
```
